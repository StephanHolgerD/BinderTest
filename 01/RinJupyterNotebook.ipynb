**Short summary of how to use the jupyter notebook:**

1. think about what operation you want to do


2. write the your code in a *Code* cell


3. execute the cell with shift + enter


4. check the result (diplayed under the cell or in the background)


5. if an error appears **read** the error message


6. correct your code



7. if you dont understand the error message --> **read the documentation** (shift + tab)


8. correct your code
